```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'Inequality in Education (1).csv'
data = pd.read_csv(url)
countries = ['United States', 'United Kingdom', 'France', 'Germany', 'Japan']
data = data[data['Country'].isin(countries)]
data_long = pd.melt(data, id_vars=['Country'], 
                    value_vars=[col for col in data.columns if 'Inequality in Education' in col],
                    var_name='Year', value_name='Inequality')

data_long['Year'] = data_long['Year'].str.extract('(\d+)').astype(int)

plt.figure(figsize=(12, 6))
sns.lineplot(data=data_long, x='Year', y='Inequality', hue='Country', marker='o')

plt.title('Inequality in Education Over Time')
plt.xlabel('Year')
plt.ylabel('Inequality in Education')
plt.xticks(data_long['Year'].unique())  # Ensure all years are shown on x-axis
plt.grid(True)
plt.legend(title='Country')
plt.show()
```

```python
url = 'Inequality in Education (1).csv'
data = pd.read_csv(url)
data = data[['Country', 'UNDP Developing Regions', 'Inequality in Education (2021)']]
data.dropna(inplace=True)
average_inequality = data.groupby('UNDP Developing Regions')['Inequality in Education (2021)'].mean().reset_index()
plt.figure(figsize=(10, 7))
plt.pie(average_inequality['Inequality in Education (2021)'], 
        labels=average_inequality['UNDP Developing Regions'], 
        autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(average_inequality)))
plt.title('Average Inequality in Education by UNDP Developing Regions (2021)')
plt.axis('equal') 
plt.show()
```

```python
url = 'Inequality in Education (1).csv'
data = pd.read_csv(url)
data = data[['Country', 'HDI Rank (2021)', 'Inequality in Education (2021)']]
data.dropna(inplace=True)
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='HDI Rank (2021)', y='Inequality in Education (2021)', hue='Country', legend=False)
plt.title('HDI vs Inequality in Education (2021)')
plt.xlabel('Human Development Index (HDI Rank)')
plt.ylabel('Inequality in Education (2021)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'Inequality in Education (1).csv'
data = pd.read_csv(file_path)

# Prepare the data for regional analysis
regional_data = data.groupby('UNDP Developing Regions')['Inequality in Education (2021)'].mean().reset_index()

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Inequality in Education (2021)', y='UNDP Developing Regions', data=regional_data, palette='viridis')

# Set the chart title and labels
plt.title('Average Inequality in Education by UNDP Developing Regions (2021)')
plt.xlabel('Average Inequality in Education (2021)')
plt.ylabel('UNDP Developing Regions')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:/Users/hp/OneDrive/Desktop/Transformed.csv")

# top 5 countries with highest education inequality in 2021
plt.figure(figsize=(10, 6))
colorsbar = ['blue', 'red', 'yellow', 'green', 'pink']

df_sorted = df.nlargest(5, 'Education_Inequality21').sort_values('Education_Inequality21', ascending=False)
sns.barplot(x='Country', y='Education_Inequality21', data=df_sorted, palette=colorsbar)  

plt.title('Top 5 Countries with Highest Education Inequality in 2021')
plt.xlabel('Country')  
plt.ylabel('Education Inequality') 
plt.xticks(rotation=45)  
plt.grid(axis='y', linestyle='--')  

plt.show()
```

```python
# top 5 countries with lowest education inequality in 2021
top_5_lowest_inequality = df[['Country', 'Education_Inequality21']].sort_values('Education_Inequality21').head(5)
colors=['orange','purple','magenta','brown','maroon']

plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Education_Inequality21', data=top_5_lowest_inequality, palette=colors)
plt.title('Top 5 Countries with Lowest Education Inequality in 2021')
plt.xlabel('Country')
plt.ylabel('Education Inequality')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.show()
```
