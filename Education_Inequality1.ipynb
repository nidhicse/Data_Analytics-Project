```python
df2= df.drop( columns=['Country','Human_Development_Groups','UNDP_Developing_Regions','HDI_Rank_21','ISO3'])
df2.head()
```

```python
df_transformed = df2.T
df_transformed
```

```python
df_transformed = df_transformed.fillna(df_transformed.mean())
```

```python
df3=df_transformed.T
```

```python
df3["Country"]=df["Country"]
```

```python
df3["HDI_Rank"]=df["HDI_Rank_21"]
```

```python
df3.to_csv("Transformed.csv",sep=",",index=False,encoding='utf-8')
```
